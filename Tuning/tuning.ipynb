```python
NCP_API_KEY = os.getenv('NCP_API_KEY')
DATASET_ACCESS_KEY = os.getenv('DATASET_ACCESS_KEY')
DATASET_SECRET_KEY = os.getenv('DATASET_SECRET_KEY')
```

```python
import requests

class CreateTaskExecutor:
    def __init__(self, host, uri, api_key, request_id):
        self._host = host
        self._uri = uri
        self._api_key = api_key
        self._request_id = request_id
    
    def _send_request(self, create_request):
        headers = {
            'Authorization': self._api_key,
            'X-NCP-CLOVASTUDIO-REQUEST-ID': self._request_id,
        }

        result = requests.post(self._host + self._uri, json=create_request, headers=headers).json()
        
        return result

    def execute(self, create_request):
        res = self._send_request(create_request)
        if 'status' in res and res['status']['code'] == '20000':
            return res['result']

        else:
            return res
```

```python
host='https://clovastudio.stream.ntruss.com'
uri='/tuning/v2/tasks'
api_key='Bearer ' + str(NCP_API_KEY)
request_id='<request_id>'

epochs = 8
bucket_name = 'instruction.dataset'
file_path = 'job_instruction_dataset.csv'
```

```python
import json 
if __name__ == "__main__":
    completion_executor = CreateTaskExecutor(host, uri, api_key, request_id)

    request_data = {
        'name': 'generation_task',
        'model': 'HCX-003',
        'tuningType': 'PEFT',
        'trainEpochs': str(epochs),        
        'learningRate': '1e-5',         
        'trainingDatasetBucket': bucket_name,
        'trainingDatasetFilePath': file_path,
        'trainingDatasetAccessKey': DATASET_ACCESS_KEY,
        'trainingDatasetSecretKey': DATASET_SECRET_KEY
    }

    response = completion_executor.execute(request_data)

    # print(request_data)
    print(json.dumps(response, indent=4, ensure_ascii=False))
    
```

```python
import time
import json
import os
import requests

class FindTaskExecutor:
    def __init__(self, host, uri, api_key, request_id):
        self._host = host
        self._uri = uri
        self._api_key = api_key
        self._request_id = request_id

    def _send_request(self, task_id):
        headers = {
            'Content-Type': 'application/json; charset=utf-8',
            'Authorization': self._api_key,
            'X-NCP-CLOVASTUDIO-REQUEST-ID': self._request_id
        }
        res = requests.get(self._host + self._uri + task_id, headers=headers)
        return res.json()

    def execute(self, task_id):
        res = self._send_request(task_id)
        if 'status' in res and res['status']['code'] == '20000':
            return res['result']
        return res


if __name__ == "__main__":
    NCP_API_KEY = os.getenv("NCP_API_KEY")

    executor = FindTaskExecutor(
        host='https://clovastudio.stream.ntruss.com',
        uri='/tuning/v2/tasks/',
        api_key='Bearer ' + str(NCP_API_KEY),
        request_id='check-training-status'
    )

    task_id = '00vpqbzj' 

    while True:
        result = executor.execute(task_id)  

        status = result.get("status")
        info = result.get("statusInfo", {}) or {}

        print(f"Task: {task_id}, status: {status}")

        curr_step    = info.get("currStep")
        total_steps  = info.get("totalTrainSteps")
        curr_epoch   = info.get("currEpoch")
        total_epoch  = info.get("totalTrainEpochs")
        train_loss   = info.get("trainLoss")
        data_rows    = info.get("dataRows")
        tokens       = info.get("numOfTokens")
        failure      = info.get("failureReason")
        message      = info.get("message")
        progress     = info.get("progressInfo", [])

        print(f"  step: {curr_step}/{total_steps}, epoch: {curr_epoch}/{total_epoch}")
        print(f"  loss: {train_loss}, rows: {data_rows}, tokens: {tokens}")
        print(f"  failure: {failure}, message: {message}")
        for p in progress:
            print(f"  - {p['step']}: {p['progress']}")

        # dng khi job xong
        if status in ("FAILED", "SUCCEEDED"):
            print("Training finished with status:", status)
            break

        time.sleep(10)
```
